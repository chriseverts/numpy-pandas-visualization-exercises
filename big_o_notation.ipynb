```python
#Make a new Jupyter notebook named big_o_notation.ipynb

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math as math


plt.figure(figsize=(14, 6)) # (width, height)
x = list(range(1,5))


y1 = [n + 1 for n in x]
y2 = [np.log(n)  for n in x]
y3 = [n for n in x]
y4 = [np.log(n) * 2 for n in x]
y5 = [n**2 for n in x]
y6 = [2**n  for n in x]
y7 = [math.factorial(n) for n in x]
y8 = [n**n for n in x]

plt.ylim(0,20)
plt.plot(x, y1, c='orange', label = '$O(1)')
plt.plot(x, y2, c='red', label = '$O(log n)')
plt.plot(x, y3, c='yellow', label = '$O(n)')
plt.plot(x, y4, c = 'green', label = '$O(n log n)$)')
plt.plot(x, y5, c='purple', label = '$O(n^2)$')
plt.plot(x, y6, c='skyblue', label = '$O(2^n)$')
plt.plot(x, y7, c='darkred', label = '$O(n!)$')
plt.plot(x, y8, c = 'b', label = '$O(n^n))')



plt.title('Big O Notation')
plt.xlabel('$Elements$')
plt.ylabel('$Operations$')
plt.legend()
plt.show()


```
