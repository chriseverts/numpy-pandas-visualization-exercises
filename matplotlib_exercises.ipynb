```python
#Use matplotlib to plot the following equation:

#y=x2−x+2

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


x = list(range(-20, 20)) # or some other range of numbers for your x values

y = [n**2 - n + 2 for n in x]
plt.plot(x,y)
plt.title('$y=x^2 - x +2$')
plt.text(0,0, '(0,0)', fontsize=10)
plt.xlabel('$x$')
plt.ylabel('y')
plt.annotate('Origin', xy=(0, 0), xytext=(-5, 100),
             arrowprops={'facecolor': 'blue'})
plt.show()


```

```python
x = list(range(-10, -1)) # or some other range of numbers for your x values

y = [1/(n+1) for n in x]
plt.plot(x,y)
plt.title('y = 1/(n+1)')

plt.xlabel('$x$')
plt.ylabel('y')
plt.show()
```

```python
plt.figure(figsize = (10,10))
plt.subplot (2,2,1)
x = list(range(0, 10)) # or some other range of numbers for your x values

y1 = [n**3 for n in x]
plt.plot(x,y1, c= 'orange')
plt.title('y = n**3')


plt.subplot (2,2,2)
x2 = list(range(0, 20)) # or some other range of numbers for your x values

y2 = [n**1/2  for n in x2]
plt.plot(x2,y2, c= 'Red')
plt.title('y = n**1/2')

plt.subplot(2,2,3)

x3 = list(range(0, 15)) # or some other range of numbers for your x values

y3 = [1/(n+1) for n in x3]
plt.plot(x3,y3, c= 'yellow')
plt.title('y = 1/(n+1)')


plt.subplot(2,2,4)
x4 = list(range(0, 5)) # or some other range of numbers for your x values

y4 = [2**n for n in x4]
plt.plot(x4,y4, c= 'green')
plt.title('y = 2**n')

plt.legend(loc='upper right')
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize = (10,5))
x = list(range(0,15))
y1 = [n**3 for n in x]
y2 = [n**1/2  for n in x2]
y3 = [1/(n+1) for n in x3]
y4 = [2**n for n in x4]
plt.plot(x, y1, c='orange', label = '$y=n**3$')
plt.plot(x2, y2, c='red', label = '$y = n**1/2$')
plt.plot(x3, y3, c='yellow', label = '$y= 1/(n+1$')
plt.plot(x4, y4, c = 'green', label = 'y = 2**n)')
plt.ylim(0,10)
plt.xlabel('x')
plt.ylabel('y')
plt.title("All graphs")
plt.legend()
plt.show()
```
