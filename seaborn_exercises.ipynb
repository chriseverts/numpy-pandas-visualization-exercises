```python
#Use the iris database to answer the following quesitons:
##What does the distribution of petal lengths look like?
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data

import seaborn as sns

iris_df = data('iris')
iris_df

iris_df = iris_df.rename(columns= {
                                    'Sepal.Length': 'sepal_length',
                                    'Sepal.Width': 'sepal_width',
                                    'Petal.Length':'petal_length',
                                    'Petal.Width': 'petal_width',
                                    'Species': 'species'
})
iris_df
```

```python

##What does the distribution of petal lengths look like?


sns.histplot(iris_df.petal_length)
```

```python
#2.) Is there a correlation between petal length and petal width?
sns.relplot(x='petal_length', y='petal_width', data=iris_df)
plt.title('Petal Length vs. Petal Width')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')

## there appears to be a correlation
```

```python
##Would it be reasonable to predict species based on sepal width and sepal length?

#3.) Would it be reasonable to predict species based on sepal width and sepal length?
sns.relplot(x='sepal_length', y='sepal_width', hue = 'species', data=iris_df)
plt.title('Sepal Length vs. Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

```

```python
##Which features would be best used to predict species?


#4.) Which features would be best used to predict species?
 #probably petal length and width. It's strongly correlated.
sns.relplot(x='petal_length', y='petal_width', hue = 'species', data=iris_df)
plt.title('Petal Length vs. Petal Width')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
```
